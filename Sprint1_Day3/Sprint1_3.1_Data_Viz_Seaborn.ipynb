# Data Visualization using Seaborn

- Seaborn is a Python visualization library based on matplotlib. 
- It provides a high-level interface for drawing - attractive statistical graphics. Here is the [documentation](http://seaborn.pydata.org/index.html).
- Go ahead and import seaborn. By convention, we alias seaborn as sns


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import streamlit as st
```

```python
df = pd.read_csv("regional_stats.csv")
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
df.shape
```

### 1. Boxplot

Let's create a simple boxplot with seaborn.

```python
plt.figure(figsize=(4,6))

sns.boxplot(y=df["NAT_MATH_PRIMARY"])
plt.title("Distribution of Scores in the National Achievement Test")

plt.show()
```

How do we read boxplots?

<img src=https://jingwen-z.github.io/images/20190319-boxplot.png width=600 height=600 />

Let's try to see if there's a difference among the scores of students in the National Achievement Test.

```python
nat = df[['NAT_MATH_PRIMARY', 'NAT_ENGLISH_PRIMARY', 'NAT_SCIENCE_PRIMARY',
          'NAT_MATH_SECONDARY', 'NAT_ENGLISH_SECONDARY', 'NAT_SCIENCE_SECONDARY']]
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(x="value", y="variable", data=pd.melt(nat))
plt.title("Distribution of Scores in the National Achievement Test")

plt.show()
```

### 2. Correlation Plot
It measures the degree to which there is a linear relationship between the two variables. It is the covariance of the two variables normalized by the variance of each variable.

```python
df.corr()
```

<img src=https://www.mathsisfun.com/data/images/correlation-examples.svg width=1000 height=1000 />

However, we have to emphasize the saying that **"correlation does not imply causation"**. You may conclude that two variable have a strong relationship or a weak relationship but correlation is not enough to be able to say that one variable causes the change to another variable. Always keep that in mind!

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(), center=0.00)

plt.show()
```

### 2. Other Plots


#### a) Pairplot

```python
plt.figure(figsize=(10,8))

sns.pairplot(df[['NET_ENROLMENT_ELEMENTARY', 'COHORT_SURVIVAL_ELEMENTARY',
                 'GROSS_ENROLLMENT_PRIMARY', 'TEACHER_PRIMARY', 'TEXTBOOKS_DISTRIBUTED']])

plt.show()
```

#### b) Clustered bar

```python
region = df.groupby("REGION")['NAT_MATH_PRIMARY', 'NAT_ENGLISH_PRIMARY', 'NAT_SCIENCE_PRIMARY',
          'NAT_MATH_SECONDARY', 'NAT_ENGLISH_SECONDARY', 'NAT_SCIENCE_SECONDARY'].mean().reset_index()
```

```python
region2 = pd.melt(region, id_vars="REGION")
region2
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x="REGION", y="value", hue="variable", data=region2)
plt.show()
```
