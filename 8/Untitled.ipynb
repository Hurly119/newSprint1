```python
teachers = pd.read_csv("Teachers data.csv")
teachers.head()
```

```python
teachers.shape
```

```python
schoolsRooms = pd.merge(schools,rooms, left_on="school.id",right_on="School ID",how="left")
schoolsRooms.head()
```

```python
schoolsRooms.drop("School ID",axis=1,inplace=True)
```

```python
schRoomTeach = pd.merge(schoolsRooms,teachers,on="school.id",how="left")
```

```python
schRoomTeach.columns
```

```python
MOOE.columns
```

```python
MOOELoc = pd.merge(MOOE,schoolsLoc,left_on="school.id",right_on="School ID",how="right")
```

```python
allSchools = pd.merge(MOOELoc,schRoomTeach,on="school.id")
```

```python
allSchools.columns
```

```python
allSchools.drop(["school.id","school.name_x","school.name_y","District","Division","Region","Province","Municipality"],inplace=True,axis=1)
```

```python
allSchools.drop(["school.abbrev","school.cityclass"],inplace=True,axis=1)
```

```python
allSchools.isna().sum()
```

```python
allSchools
```
